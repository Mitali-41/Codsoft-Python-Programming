```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

def button_click(value):
    current = entry_var.get()
    entry_var.set(current + str(value))

def clear():
    entry_var.set("")

def calculate():
    try:
        expression = entry_var.get()
        result = eval(expression)
        entry_var.set(str(result))
    except:
        messagebox.showerror("Error", "Invalid Expression")
        entry_var.set("")

root = tk.Tk()
root.title("Simple Calculator")
root.geometry("320x400")
root.configure(bg="#2e2e2e")

style = ttk.Style()
style.theme_use("clam")
style.configure("TEntry", padding=5, font=("Segoe UI", 16))
style.configure("TButton", font=("Segoe UI", 14), padding=10)
style.configure("TLabel", background="#2e2e2e", foreground="white")

entry_var = tk.StringVar()
entry = ttk.Entry(root, textvariable=entry_var, width=20, justify="right", font=("Segoe UI", 18))
entry.pack(pady=20)

button_frame = tk.Frame(root, bg="#2e2e2e")
button_frame.pack()

buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('+', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('-', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('*', 3, 3),
    ('0', 4, 0), ('.', 4, 1), ('C', 4, 2), ('/', 4, 3),
    ('=', 5, 0, 4)
]

for btn in buttons:
    text = btn[0]
    row = btn[1]
    col = btn[2]
    colspan = btn[3] if len(btn) == 4 else 1
    action = (
        calculate if text == '=' else
        clear if text == 'C' else
        lambda val=text: button_click(val)
    )
    ttk.Button(button_frame, text=text, command=action).grid(row=row, column=col, columnspan=colspan, sticky="nsew", padx=3, pady=3)

for i in range(6):
    button_frame.rowconfigure(i, weight=1)
for j in range(4):
    button_frame.columnconfigure(j, weight=1)

root.mainloop()
```
