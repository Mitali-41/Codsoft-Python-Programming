```python
import tkinter as tk
from tkinter import messagebox

def add_task():
    task = entry.get()
    if task != "":
        listbox.insert(tk.END, task)
        entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Task cannot be empty")

def delete_task():
    selected = listbox.curselection()
    if selected:
        listbox.delete(selected)

def clear_all():
    listbox.delete(0, tk.END)

def update_task():
    selected = listbox.curselection()
    new_task = entry.get()
    if selected and new_task != "":
        listbox.delete(selected)
        listbox.insert(selected, new_task)
        entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Select a task and enter new text")

def mark_done():
    selected = listbox.curselection()
    if selected:
        current = listbox.get(selected)
        if not current.startswith("✅ "):
            listbox.delete(selected)
            listbox.insert(selected, "✅ " + current)

def mark_undone():
    selected = listbox.curselection()
    if selected:
        current = listbox.get(selected)
        if current.startswith("✅ "):
            listbox.delete(selected)
            listbox.insert(selected, current[2:])

root = tk.Tk()
root.title("To-Do List")
root.geometry("320x450")

frame = tk.Frame(root)
frame.pack(pady=10)

entry = tk.Entry(frame, width=25)
entry.pack(side=tk.LEFT, padx=5)

add_button = tk.Button(frame, text="Add", width=8, command=add_task)
add_button.pack(side=tk.LEFT)

listbox = tk.Listbox(root, width=40, height=15)
listbox.pack(pady=10)

update_button = tk.Button(root, text="Update", width=10, command=update_task)
update_button.pack(pady=2)

done_button = tk.Button(root, text="Mark Done", width=10, command=mark_done)
done_button.pack(pady=2)

undone_button = tk.Button(root, text="Unmark Done", width=10, command=mark_undone)
undone_button.pack(pady=2)

delete_button = tk.Button(root, text="Delete", width=10, command=delete_task)
delete_button.pack(pady=2)

clear_button = tk.Button(root, text="Clear All", width=10, command=clear_all)
clear_button.pack(pady=2)

root.mainloop()

```
